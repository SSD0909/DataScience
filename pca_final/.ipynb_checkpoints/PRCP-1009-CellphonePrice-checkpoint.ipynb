#PREDICTION

```python


#split data into x and y
x  =  mp.drop('price_range', axis =1)
y  = mp['price_range'].values
```

```python
#get x train ,y train, x test and y test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=10)
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model   =  KNeighborsClassifier(n_neighbors=13)  #initialize model
```

```python
model.fit(x_train, y_train)#training model   # data + algorithm --> model 
y_pred = model.predict(x_test)
print(f'KNN Model Score: {model.score(x_test, y_test) * 100}%')
```

```python
##Hyper Parameter Tuning

train_scores = []

test_scores = []

neighbors= range(1,25)

for i in neighbors:
        model.set_params(n_neighbors = i)
  
        train_scores.append(model.score(x_train, y_train))
        test_scores.append(model.score(x_test, y_test))
plt.plot(neighbors, train_scores, label="Train Scores")
plt.plot(neighbors, test_scores, label="Test Scores")
plt.xticks(np.arange(1, 25,1))
plt.xlabel("No. of neighbors")
plt.ylabel("Model score")
plt.legend()

print(f"Maximum KNN score on the test data: {max(test_scores)*100:.2f}%")
```

```python
model   =  KNeighborsClassifier(n_neighbors=17)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(f'KNN Model Score: {model.score(x_test, y_test) * 100}%')
```

#Model Evaluation

```python
print(confusion_matrix(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(3, 3))
ax = sns.heatmap(confusion_matrix(y_test, y_pred),annot=True)    
plt.xlabel("True value")
plt.ylabel("Predicted value")  
```

```python
print(classification_report(y_test,y_pred))
```
