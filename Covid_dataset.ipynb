Q3. a. Are there any missing values? b. Please handle them. Write explanation of what you did to handle them.

```python
#Missing values
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
# Filling the Missing values
#Since the missing values for Low Income Area (LIA) Census Tract (Poverty Percentage) is a numerical data , i have used the mean to fill the values .
df['Low Income Area (LIA) Census Tract (Poverty Percentage)'] = df['Low Income Area (LIA) Census Tract (Poverty Percentage)'].fillna(df['Low Income Area (LIA) Census Tract (Poverty Percentage)'] .mean())


# Replace the value with 0

df['Total Score'] = df['Total Score'].replace(['Manual Verification Required'], 'Null') 


df['Tribal Community\rScore (Geographic Only)'] = df['Tribal Community\rScore (Geographic Only)'].replace(['Manual Verification Required'], 'Null') 
df.isnull().sum()
```

```python
# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Visual Exploratory Data Analysis (EDA)
sns.countplot(y='Hardest Hit Area (HHA)', data=df)
```

```python
fig, ax = plt.subplots(figsize=(10, 10)) 
sns.barplot(x='Rural - Score', y='Total Score', data=df)
```

```python
score = df[df['HHA Score']>5]
score.tail(10)
```

```python
## Creating a pivot to get the aggregate value of the hardest hit areas
#sorted to get the TOP 10 States which have Sum of hard hit scores 

pivot = pd.pivot_table(
    data=score,
    index=['STATE', 'COUNTY'],
    aggfunc={'HHA Score' : 'sum'}
)
sort = pivot.sort_values("HHA Score", axis = 0, ascending = False)

## The below observation gives the TOP 10 hard Hit area

print(sort.head(10))
```
