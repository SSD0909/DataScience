```python
# Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

# Import Breast Cancer Data

```python
df = pd.read_csv('breast_cancer.csv')
df.head()
```

# Consider 2 features and map it as x and outcome as y

# Train and Test split

```python
x  =  df.iloc[:,:2].values
y =  df.outcome.values
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.20, random_state=100)
```

# Initializing the classifier and Modify the C and Gamma value to view the plot

```python
# Try with different C value to check the plot
svm = SVC(C=200.0,kernel='rbf',gamma=0.05,verbose=True,random_state=100)

#note degree will be ignored by rbf 
```

```python
svm.fit(x_train,y_train)
```

```python
# Use plot decision region in 2 dimension

# Adding axes annotations
plot_decision_regions(x_train, y_train,clf=svm, legend =2)



```

# Prediction for RBF Kernal

```python
model_prediction =  svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score, confusion_matrix
print("accuracy score:" , accuracy_score(y_test , model_prediction))
print("precision score:" , precision_score(y_test , model_prediction))
print("recall:" , recall_score(y_test , model_prediction))
print("f1 score:" , f1_score(y_test , model_prediction))
```

# Linear SVM

```python
svm = SVC(C=200, kernel='linear')
svm.fit(x_train,y_train)
# Use plot decision region in 2 dimension
#pca = PCA(n_components = 2)
# Adding axes annotations
plot_decision_regions(x_train, y_train,clf=svm, legend =2)
```

# Prediction with Linear SVM

```python
model_prediction =  svm.predict(x_test)
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score, confusion_matrix
print("accuracy score:" , accuracy_score(y_test , model_prediction))
print("precision score:" , precision_score(y_test , model_prediction))
print("recall:" , recall_score(y_test , model_prediction))
print("f1 score:" , f1_score(y_test , model_prediction))
```

# Polynomial Kernal

```python
svm = SVC(C=200, kernel='poly')
svm.fit(x_train,y_train)
# Use plot decision region in 2 dimension
#pca = PCA(n_components = 2)
# Adding axes annotations
plot_decision_regions(x_train, y_train,clf=svm, legend =2)
# Adding axes annotations
plt.xlabel('Mean Radius [cm]')
plt.ylabel('Mean Texure [cm]')
plt.title('Breast Cancer Plot')

plt.show()
```

# Prediction with Polynomial Kernal

```python
model_prediction =  svm.predict(x_test)
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score, confusion_matrix
print("accuracy score:" , accuracy_score(y_test , model_prediction))
print("precision score:" , precision_score(y_test , model_prediction))
print("recall:" , recall_score(y_test , model_prediction))
print("f1 score:" , f1_score(y_test , model_prediction))
```

# Ploynomial Kernal with different degree

```python
for i in range(2,6):
    svm = SVC(C=100, kernel='poly', degree = i)
    svm.fit(x_train,y_train)
    # Use plot decision region in 2 dimension
    #pca = PCA(n_components = 2)
    # Adding axes annotations
    plot_decision_regions(x_train, y_train,clf=svm, legend =2)
    # Adding axes annotations
    plt.xlabel('Mean Radius [cm]')
    plt.ylabel('Mean Texure [cm]')
    plt.title('Breast Cancer Plot')
    plt.show()
    model_prediction =  svm.predict(x_test)
    from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score, confusion_matrix
    print("accuracy score:" , accuracy_score(y_test , model_prediction))
    print("precision score:" , precision_score(y_test , model_prediction))
    print("recall:" , recall_score(y_test , model_prediction))
    print("f1 score:" , f1_score(y_test , model_prediction))
```

# RBF with different gamma and constant C value

```python
gamma = [0.1, 1,10,50,100,150,200,500,750,1000]
for i in range(len(gamma)):
    svm = SVC(C=100.0,kernel='rbf',degree=3,gamma=gamma[i],verbose=True,random_state=100)
    svm.fit(x_train,y_train)
    # Use plot decision region in 2 dimension
    #pca = PCA(n_components = 2)
    # Adding axes annotations
    plot_decision_regions(x_train, y_train,clf=svm, legend =2)
      # Adding axes annotations
    plt.xlabel('Mean Radius [cm]')
    plt.ylabel('Mean Texure [cm]')
    plt.title('Breast Cancer Plot')
    plt.show()
    model_prediction =  svm.predict(x_test)
    from sklearn.metrics import accuracy_score , recall_score 
    precision_score , f1_score, confusion_matrix
    print("accuracy score:" , accuracy_score(y_test , model_prediction))
    print("precision score:" , precision_score(y_test , model_prediction))
    print("recall:" , recall_score(y_test , model_prediction))
    print("f1 score:" , f1_score(y_test , model_prediction))
```
